```python
import os
import csv
```

```python
# Define the paths to your directories
fake_videos_dirs = [
    # '/content/drive/MyDrive/Datasets/Fake_Face_Dataset/Deepfakes_Fake_Face_data'
    # '/content/drive/MyDrive/Datasets/Celeb_DF/FAKE'
    # '/content/drive/MyDrive/Datasets/DFDC_DATASET/FAKE'
    '/content/drive/MyDrive/Datasets/FineTuning_dataset/FAKE'
]

real_videos_dirs = [
    # '/content/drive/MyDrive/Datasets/Real_Face_Dataset/youtube_Real_Face_data'
    # '/content/drive/MyDrive/Datasets/Celeb_DF/REAL'
    # '/content/drive/MyDrive/Datasets/DFDC_DATASET/REAL'
    '/content/drive/MyDrive/Datasets/FineTuning_dataset/REAL'
]

output_csv = '/content/drive/MyDrive/DatasetsLabels/fineTune_video_labels.csv'

# List all video files in the fake directories
fake_videos = []
for dir in fake_videos_dirs:
    fake_videos.extend(os.listdir(dir))

# List all video files in the real directories
real_videos = []
for dir in real_videos_dirs:
    real_videos.extend(os.listdir(dir))

# Create and write to the CSV file
with open(output_csv, 'w', newline='') as csvfile:
    csvwriter = csv.writer(csvfile)
    # Write the header
    csvwriter.writerow(['filename', 'label'])

    # Write the fake video files with the label 'fake'
    for video in fake_videos:
        csvwriter.writerow([video, 'fake'])

    # Write the real video files with the label 'real'
    for video in real_videos:
        csvwriter.writerow([video, 'real'])

print(f"CSV file '{output_csv}' created successfully.")
```
